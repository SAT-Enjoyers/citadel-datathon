```python
# Let's first read the uploaded CSV file to understand its structure and contents.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the uploaded CSV file
file_path = '../dataset/Meat_Stats_Slaughter_Weights.csv'
df = pd.read_csv(file_path)

df = df[df['Animal'] == 'Cattle']
```

```python
# Convert 'Weight' column to numeric, removing commas
df['Weight'] = pd.to_numeric(df['Weight'].str.replace(',', ''), errors='coerce')

# Create a datetime column from 'Year' and 'Month'
df['Datetime'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')

# Plotting
plt.figure(figsize=(14, 8))

# Get unique animals
animals = df['Animal'].unique()

for animal in animals:
    subset = df[df['Animal'] == animal]
    plt.plot(subset['Datetime'], subset['Weight'], label=animal)

plt.title('Weight Over Time for Different Animals')
plt.xlabel('Date')
plt.ylabel('Weight in Pounds')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
